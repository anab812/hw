```
import numpy as np
from scipy.optimize import differential_evolution
```

```
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

result = differential_evolution(ackley, bounds, seed=42)
result
```

strategy

```
strategies = ['best1bin', 'best1exp','rand1exp','randtobest1exp','currenttobest1exp','best2exp','rand2exp','randtobest1bin','currenttobest1bin','best2bin','rand2bin','rand1bin']
```

```
output_strategy = {}
```

```
for s in strategies:
  result = differential_evolution(ackley, bounds, strategy = s, seed=42)
  output_strategy[s] = dict(result)['nit']
print(output_strategy)
```

```
strategy_list = sorted(output_strategy.items()) 

x, y = zip(*strategy_list)

plt.bar(x, y, )
fig = plt.gcf()
fig.set_size_inches(25, 15)
plt.xlabel("strategy", fontsize=16)
plt.ylabel("number of iteration", fontsize=16)
plt.show()
```
