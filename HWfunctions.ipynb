```
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def init_p(documents):
  doc = str(input('Введите номер документа: '))
  for document in documents:
    for v in document.values():
      if doc == str(v):
          return (f'Владелец документа: {document["name"]}')
      else:
          return ('Документ не найден в базе')
# print(init_p(documents)) 

def init_s(documents, directories):
  shelf = input('Введите номер документа: ')
  for k, d in directories.items(): 
    for i in d:
      if shelf == i:
         return (f'Документ хранится на полке: {k}')
    else:
      return ('Документ не найден в базе')
# print(init_s(documents, directories))    
    
def init_l(documents, directories):

 for document in documents:
     data = ['№:', document["number"], ', тип: ', document["type"], ', владелец:', document["name"], ', полка хранения: ']
     for s, k in directories.items():
        for n in k:
          if n == data[1]:
            data.append(s) 
     print (''.join(data))
# print(init_l(documents, directories))

def init_ads(documents, directories):
  shelf = input('Введите номер полки: ')
  if shelf not in directories.keys():
     directories[shelf] = []
     return (f'Полка добавлена. Текущий перечень полок: {list(directories.keys())}.')
  else: 
     print (f'Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}.')
# print(init_ads(documents, directories))

def init_ds(documents, directories):
    shelf = input('Введите номер полки: ')
    # for v in directories.values():
    if shelf not in list(directories.keys()):
        return (f'Такой полки не существует. Текущий перечень полок: {list(directories.keys())}.')
    elif len(directories[shelf]) == 0:
        directories.pop(shelf)
        return (f'Полка удалена. Текущий перечень полок: {list(directories.keys())}.')
    else:
        return (f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list(directories.keys())}.')

# print(init_ds(documents, directories))   

def main(documents):
    while True:
        user_input = input('Введите команду ')
        if user_input == 'p':
            print(init_p(documents))
        elif user_input == 's':
            print(init_s(documents, directories))
        elif user_input == 'l':
            print(init_l(documents, directories))
        elif user_input == 'ads':
            print(init_ads(documents, directories))
        elif user_input == 'ds':
            print(init_ds(documents, directories))
        elif user_input == 'q':
            break
        else:
            print('Введена неверная команда')


main(documents)
```
